"""
Create a comprehensive unit converter that handles:

1. Length conversions:
   - Meters to/from Feet
   - Kilometers to/from Miles
   - Centimeters to/from Inches

2. Weight conversions:
   - Kilograms to/from Pounds
   - Grams to/from Ounces

3. Volume conversions:
   - Liters to/from Gallons
   - Milliliters to/from Fluid Ounces

Requirements:
- Use a menu system to select conversion type
- Handle invalid inputs gracefully
- Format output to 2 decimal places
- Add helpful descriptions (e.g., "That's about the weight of a small car!")
- Use functions for each conversion type
- Include input validation (no negative values for these units)

Bonus challenges:
- Add temperature conversions (C/F/K)
- Allow continuous conversions until user chooses to exit
- Store conversion history and display at the end
"""

```python
import re

# For storing all conversions
conversion_history = []  
```

# Output Handling

```python
#Handle Float inputs
def check_positive_float(message, can_be_negative=False):
    while True:
        user_input = input(message)
        if re.fullmatch(r'-?\d+(\.\d+)?', user_input):
            value = float(user_input)
            if not can_be_negative and value < 0:
                print("Value cannot be negative. Try again.")
            else:
                return value
        else:
            print("Invalid input. Enter a number.")
```

```python
#Handle Int inputs
def get_int_choice(prompt, max_choice):
    """User chooses an integer from 1 to max_choice"""
    while True:
        user_input = input(prompt)
        if user_input.isdigit() or (user_input.startswith('-') and user_input[1:].isdigit()):
            choice = int(user_input)
            if 1 <= choice <= max_choice:
                return choice
            else:
                print(f"You must choose a number from 1 to {max_choice}.")
        else:
            print("Invalid input. Enter a number.")
```

# Length Comments and Conversion

```python
#For Adding Length helpful comments
def length_comment(value, unit="m"):
    if unit == "cm":
        value /= 100
    elif unit == "ft":
        value /= 3.28084
    elif unit == "in":
        value /= 39.3701
    elif unit == "km":
        value *= 1000
    elif unit == "mi":
        value *= 1609.34
        
    # Display a comment based on the value range
    if value < 2.5:
        return "That's about the height of a person!"
    elif value < 5:
        return "That's about the height of a room!"
    elif value < 20:
        return "That's a very tall building!"
    else:
        return "That's extremely long!"
```

```python
#For Length Conversion
def length_conversion():
    print("\n--- Length Conversions ---")
    options = ["Meters -> Feet", "Feet -> Meters", "Kilometers -> Miles",
               "Miles -> Kilometers", "Centimeters -> Inches", "Inches -> Centimeters"]
    for i, opt in enumerate(options, 1):
        print(f"{i}. {opt}")

    conv_num = get_int_choice("Enter Your Choice: ", len(options))
    value = check_positive_float("Enter The Value: ")

    #Checking Conversion Type
    if conv_num == 1:
        result = value * 3.28084
        conv_desc = f"{value} m -> {result:.2f} ft"
        unit = "m"
    elif conv_num == 2:
        result = value / 3.28084
        conv_desc = f"{value} ft -> {result:.2f} m"
        unit = "ft"
    elif conv_num == 3:
        result = value * 0.621371
        conv_desc = f"{value} km -> {result:.2f} mi"
        unit = "km"
    elif conv_num == 4:
        result = value * 1.60934
        conv_desc = f"{value} mi -> {result:.2f} km"
        unit = "mi"
    elif conv_num == 5:
        result = value / 2.54
        conv_desc = f"{value} cm -> {result:.2f} in"
        unit = "cm"
    elif conv_num == 6:
        result = value * 2.54
        conv_desc = f"{value} in -> {result:.2f} cm"
        unit = "in"

    print(f"Result = {result:.2f}")
    print(length_comment(value, unit))
    conversion_history.append(conv_desc)
```

# Weight Comments and Conversion

```python
#For Adding Weight Helpful Comments
def weight_comment(value, unit="kg"):
    if unit == "g":
        value /= 1000
    elif unit == "lb":
        value /= 2.20462
    elif unit == "oz":
        value /= 35.274
        
    # Display a comment based on the value range
    if value < 1:
        return "That's very light!"
    elif value < 50:
        return "That's a moderate weight!"
    elif value < 1000:
        return "That's about the weight of a small car!"
    else:
        return "That's extremely heavy!"
```

```python
#For Weight Conversion
def weight_conversion():
    print("\n--- Weight Conversions ---")
    options = ["Kilograms -> Pounds", "Pounds -> Kilograms",
               "Grams -> Ounces", "Ounces -> Grams"]
    for i, opt in enumerate(options, 1):
        print(f"{i}. {opt}")

    conv_num = get_int_choice("Enter Your Choice: ", len(options))
    value = check_positive_float("Enter The Value: ")

    #Checking Conversion Type
    if conv_num == 1:
        result = value * 2.20462
        conv_desc = f"{value} kg -> {result:.2f} lb"
        unit = "kg"
    elif conv_num == 2:
        result = value / 2.20462
        conv_desc = f"{value} lb -> {result:.2f} kg"
        unit = "lb"
    elif conv_num == 3:
        result = value * 0.035274
        conv_desc = f"{value} g -> {result:.2f} oz"
        unit = "g"
    elif conv_num == 4:
        result = value * 28.3495
        conv_desc = f"{value} oz -> {result:.2f} g"
        unit = "oz"

    print(f"Result = {result:.2f}")
    print(weight_comment(value, unit))
    conversion_history.append(conv_desc)
```

# Volume Comments and Conversion

```python
#For Adding Volume Helpful Comments
def volume_comment(value, unit="L"):
    if unit == "mL":
        value /= 1000
    elif unit == "gal":
        value *= 3.78541
    elif unit == "fl oz":
        value *= 0.0295735
        
    # Display a comment based on the value range
    if value < 1:
        return "That's a small bottle!"
    elif value < 10:
        return "That's about the size of a bathtub!"
    elif value < 1000:
        return "That's about the size of a small room!"
    else:
        return "That's a huge container!"
```

```python
#For Volume Conversion
def volume_conversion():
    print("\n--- Volume Conversions ---")
    options = ["Liters -> Gallons", "Gallons -> Liters",
               "Milliliters -> Fluid Ounces", "Fluid Ounces -> Milliliters"]
    for i, opt in enumerate(options, 1):
        print(f"{i}. {opt}")

    conv_num = get_int_choice("Enter Your Choice: ", len(options))
    value = check_positive_float("Enter The Value: ")
    
    #Checking Conversion Type
    if conv_num == 1:
        result = value * 0.264172
        conv_desc = f"{value} L -> {result:.2f} gal"
        unit = "L"
    elif conv_num == 2:
        result = value * 3.78541
        conv_desc = f"{value} gal -> {result:.2f} L"
        unit = "gal"
    elif conv_num == 3:
        result = value * 0.033814
        conv_desc = f"{value} mL -> {result:.2f} fl oz"
        unit = "mL"
    elif conv_num == 4:
        result = value * 29.5735
        conv_desc = f"{value} fl oz -> {result:.2f} mL"
        unit = "fl oz"

    print(f"Result = {result:.2f}")
    print(volume_comment(value, unit))
    conversion_history.append(conv_desc)
```

# Temperature Comments and Conversion

```python
#For Adding Temperature Helpful Comments
def temperature_comment(value, unit="C"):
    if unit == "F":
        value = (value - 32) * 5/9
    elif unit == "K":
        value = value - 273.15
        
    # Display a comment based on the value range
    if value < 0:
        return "Below freezing!"
    elif value < 30:
        return "Moderate temperature."
    elif value < 50:
        return "It's hot!"
    else:
        return "Extreme heat!"
```

```python
#For Temperature Conversion
def temperature_conversion():
    print("\n--- Temperature Conversions ---")
    options = ["Celsius -> Fahrenheit", "Fahrenheit -> Celsius",
               "Celsius -> Kelvin", "Kelvin -> Celsius",
               "Fahrenheit -> Kelvin", "Kelvin -> Fahrenheit"]
    for i, opt in enumerate(options, 1):
        print(f"{i}. {opt}")

    conv_num = get_int_choice("Enter Your Choice: ", len(options))
    value = check_positive_float("Enter The Value: ", can_be_negative=True)

    #Checking Conversion Type
    if conv_num == 1:
        result = (value * 9/5) + 32
        conv_desc = f"{value} °C -> {result:.2f} °F"
        unit = "C"
    elif conv_num == 2:
        result = (value - 32) * 5/9
        conv_desc = f"{value} °F -> {result:.2f} °C"
        unit = "F"
    elif conv_num == 3:
        result = value + 273.15
        conv_desc = f"{value} °C -> {result:.2f} K"
        unit = "C"
    elif conv_num == 4:
        result = value - 273.15
        conv_desc = f"{value} K -> {result:.2f} °C"
        unit = "K"
    elif conv_num == 5:
        result = (value - 32) * 5/9 + 273.15
        conv_desc = f"{value} °F -> {result:.2f} K"
        unit = "F"
    elif conv_num == 6:
        result = (value - 273.15) * 9/5 + 32
        conv_desc = f"{value} K -> {result:.2f} °F"
        unit = "K"

    print(f"Result = {result:.2f}")
    print(temperature_comment(value, unit))
    conversion_history.append(conv_desc)
```

# Main Program Loop

```python
#Main Function and Menu
def unit_converter():
    while True:
        print("\n=== Comprehensive Unit Converter ===")
        options = ["Length Conversions", "Weight Conversions",
                   "Volume Conversions", "Temperature Conversions", "Exit"]
        for i, opt in enumerate(options, 1):
            print(f"{i}. {opt}")

        conv_type = get_int_choice("Enter Your Choice From 1-5: ", len(options))

        #Checking Conversion Type
        if conv_type == 1:
            length_conversion()
        elif conv_type == 2:
            weight_conversion()
        elif conv_type == 3:
            volume_conversion()
        elif conv_type == 4:
            temperature_conversion()
        elif conv_type == 5:
            print("\n--- Conversion Report ---")
            if conversion_history:
                for i, conv in enumerate(conversion_history, 1):
                    print(f"{i}. {conv}")
            else:
                print("No conversions performed.")
            print("Thank you for using our unit converter!")

            break

# Run the program
unit_converter()
```
